# 원뿔의 부피 & 겉넓이 계산 프로그램
> 부피와 겉넓이를 구하기 위해서 알고리즘이 필요!!

<br>

부피 = $\frac{1}{3}\pi r^2 h$  
겉넓이 = $\pi r^2 + \pi rh$

![원뿔](https://drive.google.com/uc?export=view&id=1sTvE6YvL6XBU6A8IMySlUFXJ1uoUd3s-)

````python
# 반지름 : 20, 높이 : 30
# 부피
print(1 / 3 * 3.14 * 20 ** 2 * 30)
# 겉넓이
print(3.14 * 20 ** 2 + 3.14 * 20 * 30)
````

*만약 반지름과 높이를 변경하려면 위의 코드가 불편하다*

### 변수(Variable)
- 명령형 패러다임 언어의 특징
  - 처리할 데이터와 처리된 결과를 임시적 저장
  - 변수의 값을 변경하는 할당 연산자(=)를 이용

`rad = 20`
  - rad : lvalue(지속되는 대상)
  - 20 : rvalue(임시적인 대상)

rad와 20이 같다(x)  
변수 rad에 20을 할당(o)

### 값의 할당
- 프로그램이 실행되는 과정에서 처리되는 값이 어떤 행위(입력, 연산 등)에 따라 그 값이 변할 수 있는 메모리 내의 저장 공간 지정

### 식별자
> 고유하게 만들어져야 하는 저장 공간

프로그램 내부에 정의된 객체(변수, 함수 등)의 이름
  - 문자, 숫자, 밑줄로 구성
  - 문자 또는 밑줄로만 시작 가능
  - 예약어와 동일할 수 없음
  - 길이 제한이 없음

- 사용 가능한 식별자
  - volume, Surface, _50, x, y, __name
- 사용 불가한 식별자
  - \$D\$, 1-a, 2023y, else, class, r+h, 77#R
  - else, class는 예약어

### 예약어(reserved word)
- 파이썬 인터프리터에 의해 이미 문법적인 용도로 사용되어 식별자로 사용이 불가능한 단어

|---|---|---|---|---|---|
| break | except | in | True | class | finally |
| is | return | and | continue | for | lambda |
| try | as | def | global | not | with |
| async | elif | if | yield | raise | or |

*예약어를 모두 외울 필요 없이 식별자를 썼는데 계속 에러가 나면 '예약어구나...'라고 알면 됨*


````python
# 반지름(r), 높이(h) 값 할당
r = 20
h = 30

# 위의 r, h : lvalue(값 할당)
# 아래의 r, h : rvalue(값 호출)

# 부피 출력
print(1 / 3 * 3.14 * r ** 2 * h)
# 겉넓이 출력
print(3.14 * r ** 2 + 3.14 * r * h)
````

  - 1972년, 컴퓨터 분야의 노벨상인 튜링상 수상
- 프로그램 실행 흐름이 간결하고 작은 규모로 조직화하기 쉬움

### 원뿔 계산기 개선하기
> 반지름의 값이 음수가 들어온다면?  

*정확한 사용자 입력을 위해 사용자가 **잘못된 값**을 입력하면 **실행되지 않도록** 실행의 흐름을 변경해야한다.*

-> 선택구조 if를 사용하자!!
- if문 : 조건이 맞을 때만 실행

````python
#반지름 사용자 입력
rad = int(input("반지름을 입력하세요: "))
# 높이 사용자 입력
hei = int(input("높이를 입력하세요: "))

if rad > 0 :
    # 부피 & 겉넓이 계산
    vol = 1/3 * 3.14 * rad ** 2 * hei
    suf = 3.14 * rad ** 2 + 3.14 * rad * hei
    print("원뿔의 부피는 ", vol, "입니다.", sep="")
    print("원뿔의 겉넓이는 ", suf, "입니다.", sep='')
````

*반지름이 음수인 경우는 겉넓이와 부피를 계산하지 않지만  
높이가 음수라면 어떻게 될까?*

\>\> 계산을 한다.

### 논리 연산자
> 참, 거짓을 구별할 수 있는 명제를 대상으로 명제의 집합을 위해 고안한 연산자

- 두 개의 피연산자를 갖는 이항 연산자

### and(논리곱)

| b1 | b2 | b1 and b2 |
|-|-|-|
| False | False | False |
| False | True | False |
| True | False | False |
| True | True | True |

### or(논리합)

| b1 | b2 | b1 and b2 |
|-|-|-|
| False | False | False |
| False | True | True |
| True | False | True |
| True | True | True |

### not(논리부정)
> 단항 연산자로 논리값을 반전

| b1 | not b1 |
|-|-|
| False | True |
| True | False |

### 단락 평가
> Short-circuit evaluation  
> 첫 번째 논리값만으로 전체 연산 결과가 판별 가능할 때 두번째 논리값은 확인(평가)하지 않는 기법

````python
temp = 20
fruit = "apple"

print(temp >= 27 and fruit == "pear")
print(not temp <= 27)
````

#### **원뿔 계산 프로그램 개선**
> 반지름과 높이가 (10, 1), (20, 5), (30, 14)인 원뿔의 부피와 겉넓이를 각각 출력하시오.

```txt
반지름 10 , 높이 1 원뿔
부피: 104.66666666666666 넓이: 345.4
반지름 20 , 높이 5 원뿔
부피: 2093.333333333333 넓이: 1570.0
반지름 30 , 높이 14 원뿔
부피: 13188.0 넓이: 4144.8
```

반지름들 사이에 나름의 규칙이 있는 경우, 자동화할 수 있다.

````python
# 강의 실습
# 목표: range 함수를 이용하여 원뿔 계산 문제 풀기
rad_list = range(10, 31, 10)
hei_list = [1, 5, 14]

for rad, hei in zip(rad_list, hei_list):
    # 부피 & 겉넓이 계산
    vol = 1/3 * 3.14 * rad ** 2 * hei
    surf = 3.14 * rad ** 2 + 3.14 * rad * hei
    print("반지름", rad , "높이", hei, "원뿔")
    print("부피:", vol, "넓이:", surf)
````

### format 함수
> 문서를 형식화 시켜주는데 사용하는 함수  

👉 데이터를 양식에 맞춰 형식화  
![format 함수](https://drive.google.com/uc?export=view&id=1j1Bgh3OstNCb2wOKm0TxqUeTXOKb5rRX)

→ 형식 지정자: (정렬 방향, 필드 폭, 데이터 타입) 순으로 작성
- 정렬 방향
    - 왼쪽 정렬: **\<**
    - 가운데 정렬: **^**
    - 오른쪽 정렬: **\>**
- 10 : 필드 폭
- 데이터 타입
    - **d**: 정수
    - **s**: 문자열

👉 문자열 형식  
cf. **문자 타입**은 **왼쪽 정렬**이 기본
```python
format("구구단표", ">10s")
```
|←————10————→|  
□□□□□□구구단표

<br>

👉 정수 형식  
cf. **정수 타입**은 **우측 정렬**이 기본
```python
format(1234, "<10d")
```
|←———10———→|  
1234□□□□□□




````python
print(format("구구단표", ">20s"))

# 표 머리 출력
print("  |", end="")

for i in range(1, 10):
    print(format(i, ">4d"), end="")

# 새로운 행 삽입
print()
print("-"*40)

# 구구단표 출력
for i in range(1, 10, 1):
    print(i, "|", end="")
    for j in range(1, 10, ):
        print(format(i*j, "4d"), end="")
    print()
````
