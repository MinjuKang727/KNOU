```python
# rad와 hei가 모두 양수일 때, 부피&겉넓이 계산

# 반지름 사용자 입력
rad = int(input("반지름을 입력하세요. "))  # 반지름을 입력하세요. 10
# 높이 사용자 입력
hei = int(input("높이를 입력하세요. "))  # 높이를 입력하세요. -30

# 논리 연산자(3종류) : and, or, not
if rad > 0 and hei > 0 :
    # 부피 & 겉넓이 계산
    vol = 1/3 * 3.14 * rad ** 2 * hei
    suf = 3.14 * rad ** 2 + 3.14 * rad * hei
    print("부피와 겉넓이는", vol, suf, sep=' ')
else :
    print("반지름과 높이 값에 양수 값을 입력하세요.")
```

```python
# 소숫점 자릿수 제한하기

# 반지름 사용자 입력
rad = int(input("반지름을 입력하세요. "))  # 반지름을 입력하세요. 10
# 높이 사용자 입력
hei = int(input("높이를 입력하세요. "))  # 높이를 입력하세요. -30

# 논리 연산자(3종류) : and, or, not
if rad > 0 and hei > 0 :
    # 부피 & 겉넓이 계산
    vol = 1/3 * 3.14 * rad ** 2 * hei
    suf = 3.14 * rad ** 2 + 3.14 * rad * hei
    print("부피와 겉넓이는", format(vol, '>10.5f'), suf, sep=' ')
else :
    print("반지름과 높이 값에 양수 값을 입력하세요.")
```

```python
# 입력값이 조건을 만족할 때까지 반복하기
# while문

# 반지름 사용자 입력
rad = int(input("반지름을 입력하세요. "))  # 반지름을 입력하세요. 10
while rad < 0 :
    rad = int(input("반지름을 양수로 입력하세요. ")) 

# 높이 사용자 입력
hei = int(input("높이를 입력하세요. "))  # 높이를 입력하세요. -30
while hei < 0 :
    hei = int(input("높이를 양수로 입력하세요. ")) 

# 논리 연산자(3종류) : and, or, not
if rad > 0 and hei > 0 :
    # 부피 & 겉넓이 계산
    vol = 1/3 * 3.14 * rad ** 2 * hei
    suf = 3.14 * rad ** 2 + 3.14 * rad * hei
    print("부피와 겉넓이는", format(vol, '>10.5f'), suf, sep=' ')
else :
    print("반지름과 높이 값에 양수 값을 입력하세요.")
```
