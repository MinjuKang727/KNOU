[문제2] 정기예금에 대한 원금과 연이율을 입력받고 아래의 수식에 따라 월이자를 계산하여 출력하는 프로그램을 작성하시오.(소스코드는 텍스트, 실행결과는 캡처 이미지로 제출)

image.png

```python
ori = int(input("원금(원)을 입력하세요."))
rate = float(input("연이율(%)을 입력하세요."))
month_rate = ori * (rate / 1200)
print(f'월이자는 {month_rate:>10.5f}원입니다.')
```
