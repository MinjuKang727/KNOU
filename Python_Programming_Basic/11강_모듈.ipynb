````python
# import math
from math import sin

class Triangle :
    def __init__(self, a, b, alpha) :
        self.__a = a
        self.__b = b
        self.__alpha = alpha
    def get_surf(self) :
        # return 1/2 * self.__a * self.__b * math.sin(math.radians(self.__alpha))
        return 1/2 * self.__a * self.__b * sin(math.radians(self.__alpha))

tri = Triangle(10, 20, 60)
print(tri.get_surf())
````
