# 12강. 파일

## 01. 파일의 이해

### 🔷 파일의 역할
👉 컴퓨터에 의해 처리될 또는 처리된 데이터와 정보가 임시적으로 저장된 상태<br>
→ 일련의 연속된 바이트<br>
→ 프로그램(파이썬 소스코드)에 읽혀 가공·처리<br>

| 데이터 파일 | >> | **자동화**할 도구<br>처리 및 가공 | >> | 정보 파일 |
|---|---|---|---|---|

<br>

### 🔷 파일의 구성
👉 연속된 **바이트**로 구성되어 있으며<br>
연속된 바이트와 파일의 시작, 파일 포인터, 파일의 끝(EoF)으로 표현

| $바이트_0$ | $바이트_1$ | $바이트_2$ | $바이트_3$ | ⋯ | $바이트_{n-1}$ | $바이트_n$ |
|:---|:---:|---:|:---:|:---:|:---:|---:|
|↑<br><br>파일의 시작 ||↑<br><br>작업 위치<br>(파일 포인터)||||↑<br><br>파일의 끝<br>(EoF)|

<br>

### 🔷 파일의 종류
👉 데이터가 **저장되는 방식**에 따라 구분(2가지)
- 텍스트 파일 : 텍스트 에디터(혹은 리눅스의 VI 에디터)로 열어 내용을 볼 수 있음
    - 예: TXT
    - 숫자(199) 저장하기
        - 1, 9, 9 라는 서로 다른 문자로 인식
        - 별도의 인코딩 시스템(ex. ASCII)을 사용하여 각각의 문자를 바이트로 기록
        - <table>
            <tr>
                <th>199</th>
                <th>>></th>
                <th>(Text Encoding)</th>
                <th>>></th>
                <th>00111001<br>00111001<br>00110001</th>
                <th> ← 9<br> ← 9<br> ← 1</th>
            </tr>
        </table>
    - 각각은 8비트의 숫자로 이루어짐


- 바이너리 파일 : 텍스트 에디터로 열리지 않음
    - JPG, BMP, PDF 등
    - 숫자(199) 저장하기
        - 숫자의 각 자릿수를 개별적인 문자로 인식하는 것이 아니라 별도의 인코딩 시스템 없이 **실제 바이너리 숫자**로 저장
        - 한 바이트가 보통 일반적으로 0 ~ 255 혹은 -127 ~ 128까지 저장 가능한데 여기에 맞춰 한 바이트 혹은 두 바이트로 저장
        - <table>
            <tr>
                <th>199</th>
                <th>>></th>
                <th>(Binary Encoding)</th>
                <th>>></th>
                <th>11000111</th>
                <th> ← 199</th>
            </tr>
        </table>
        - 훨씬 **적은 데이터 용량**으로 실제 데이터 값 저장 가능

<u>우리가 사용하는 대다수의 컨텐츠를 좀 더 축약해서 함축적으로 저장하기 위해 **바이너리 파일**을 많이 이용</u>

<br>

### 🔷 파일 함수

👉 **파일의 시작, 파일 포인터, 파일의 끝**을 활용하여 데이터 읽기, 쓰기를 위한 함수 및 메소드를 내장

- **객체지향** : 대규모의 프로그램을 유용하게 사용할 수 있는 새로운 패러다임

|멤버|설명|생애주기(Life Cycle)|
|---|---|:---:|
|open(): file obj|파일과 연결되어 있는 파일 객체 생성|**생성**|
|read()|특정 개수의 문자를 반환|**사용**|
|readline()|한 라인의 문자열을 반환|‘‘|
|readlines(): list|전체 라인의 문자열을 리스트로 반환|‘‘|
|write(s: str)| 파일에 문자열을 작성|‘‘|
|close()|파일 닫기 및 파일 객체 삭제|**삭제**|

❓ **객체**가 **프로그램 내부에서 사용되는 과정**에 대해 고민해보자.
> **객체**의 **Life Cycle**(생애주기) : **탄생**, **발전**, **성장**, **소멸**  
> (**생성-사용-삭제**)

**파이썬 인터프리터**는 참조변수로 쓰이지 않는 객체들을 **주기적으로 점검하고 정리**함
- 참조 변수로 쓰이지 않는 객체들은 접근 불가  
    - 참조 변수로 쓰이지 않는 객체 → 사용하지 않는 객체 → 폐기된 객체
- **Garbage Collector**(쓰레기 수거기) : 사용하지 않는 객체를 별도로 모아서 삭제하는 장치
    - 인터프리터에서 주기적으로 작동

<br>

**파일**도 완벽히 **동일한 과정**(Life Cycle)을 거침  
> 파일 객체 **생성-사용-삭제**

<br>

### 🔷 파일 객체 생성
👉 구문 형식  

```python
파일객체_참조변수 = open(파일이름, 모드)
```

<u>**open함수를 사용**하는 것 빼고는 우리가 배웠던 **객체 사용법**과 거의 동일</u>

→ 물리적인 파일과 연결된 파일 객체를 생성하고 참조 변수에 할당

|모드|설명|
|---|---|
|'r'|읽기 용도|
|'w'|새로운 파일을 쓰기 용도|
|'a'|파일의 끝에 데이터를 덧붙이기 용도|

*파일의 종류에는 2가지(텍스트 파일, 바이너리 파일)가 있었는데 일단은 텍스트 파일을 다루는 모드만 배우고자 함.*

### 🔷 파일 읽기
👉 특정 범위의 데이터를 파일에서 읽고 문자열로 반환

→ **파일 포인터**(*파일을 여기까지 읽었어!*)의 이동을 동반

```text
Hamlet\nby William Shakespeare\nEdited by Ba
```

`\n` → New Line Character : 한 줄을 밑으로 내려라

```python
h_fp = open("Hamlet_by_Shakespeare.txt", "r")
# 경로가 없는 걸로 보아 같은 폴더 내에 있다는 뜻
# 읽기 모드 사용
```

-  

    ```text
    Hamlet\nby William Shakespeare\nEdited by Ba
↑
파일 포인터
    ```

```python
h_fp = open("Hamlet_by_Shakespeare.txt", "r")
title = h_fp.read(6)  # Hamlet
```
-  

    ```text
    Hamlet\nby William Shakespeare\nEdited by Ba
         ↑
         파일 포인터
    ```

```python
h_fp = open("Hamlet_by_Shakespeare.txt", "r")
title = h_fp.read(6)
author = h_fp.readline()  # by William Shakespeare
```
-  

    ```text
    Hamlet\nby William Shakespeare\nEdited by Ba
                                ↑
                                파일 포인터
    ```

<u>open()하고 사용했으면 반드시 파일 포인터를 닫아서 없애주어야 한다.</u>    
```python
h_fp = open("Hamlet_by_Shakespeare.txt", "r")
title = h_fp.read(6)
author = h_fp.readline()
h_fp.close()
```
- 읽기 모드로 파일을 열었을 때에는 파일 포인터를 닫은 후에 파일이 그대로 유지된다.

    ```text
    Hamlet\nby William Shakespeare\nEdited by Ba
    ```



<br>

❓ 파일을 읽고 난 후에는 꼭 저렇게 **close()**를 해야 하나요?<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;파일 포인터는 나중에 Garbage Collector가 없애주어도 되잖아요?
> 파일은 실제 물리적인 저장장치에 접근하여 데이터를 가공하는 과정이라고 할 수 있기때문에<br>
(*항상 그런 것은 아니지만*) close()를 정상적으로 해주지 않으면 <br>
파일 포인터에 의해 잘못된 연산이 이루어져 실제 다른 데이터가 작성된다거나 파일이 깨질 수 있으니 <br>
<u>반드시 **close()**를 해주어야 합니다.</u>


<br>

### 🔷 파일 쓰기
👉 문자열을 파일 포인터가 위치한 지점에 기록<br>
**→ w 모드로 존재하는 파일 오픈 시, 데이터 삭제**<br>

- w 모드의 특징
    - 무조건 **이름을 새로 생성**하고 **파일 포인터**를 맨 앞에 위치시킴
    - 만약 이미 존재하는 파일 이름으로 파일 생성 시, **기존 파일의 내용**이 모두 사라질 수 있다.

<u>**w 모드**를 사용할 때에는 **실제로 경로에 이미 존재**하는 파일인지 꼭 확인!!</u><br>
**덮어씌워도 괜찮을 때**에만 w 모드를 사용하세요!

```python
p_fp = open("python.txt", "w")
```
-  

    ```text

    ↑
    파일 포인터
    ```

```python
p_fp = open("python.txt", "w")
p_fp.write("KNOU\n")
```
-  

    ```text
    KNOU\n
         ↑
         파일 포인터
    ```

```python
p_fp = open("python.txt", "w")
```

-

    ```text
    KNOU\npython programming\n
    ```

```python
p_fp = open("python.txt", "w")
p_fp.write("KNOU\n")

```
- 파일 객체 삭제(close()) 시, 파일 포인터가 사라지며 실제 해당 파일이 생성됨.

    ```text
    KNOU\npython programming\n
    ```

### 🔷 데이터 추가
👉 파일의 끝에 데이터를 덧붙이는 작업<br>
→ 파일 오픈 후 파일 포인터를 EoF로 이동<br>
→ 존재하지 않는 파일은 write와 동일<br>

**a 모드** 혹은 **append 모드**라고 부름

- python.txt 파일
```text
KNOU\npython programming\n
```

1. 파일 열기
```python
a_fp = open("python.txt", "a")
```
    - 파일을 **append 모드**로 열면, 파일을 열자마자 **파일 포인터를 맨 뒤로** 옮김

    ```text
    KNOU\npython programming\n
                             ↑
                             파일 포인터
    ```

2. 데이터 추가
```python
a_fp = open("python.txt", "a")
a_fp.write("\nby CS\n")
```
    -

    ```text
    KNOU\npython programming\nby CS\n
                                    ↑
                                    파일 포인터
    ```

3. 파일 객체 삭제(close())
    - 기존 파일에 새로운 내용을 덧붙임
    ```text
    KNOU\npython programming\nby CS\n
    ```
